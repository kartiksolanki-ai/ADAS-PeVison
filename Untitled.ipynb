```python
import cv2
import numpy as np
import pygame
import math
from ultralytics import YOLO

# Initialize pygame for sound
pygame.init()
beep_sound = pygame.mixer.Sound("beep.wav")

# Load YOLOv8 model
model = YOLO("yolov8m.pt")
cap = cv2.VideoCapture("data1.mp4")

# Tracking and state management
prev_positions = {}
beep_active = False
frame_counter = 0
BEEP_THRESHOLD = 400  # Threshold for triggering the beep
kalman_filters = {}
max_vehicles_to_track = 4  # Maximum number of vehicles to track

def smooth_angles(angles, window_size=5):
    smoothed_angles = []
    for i in range(len(angles)):
        start_idx = max(0, i - window_size // 2)
        end_idx = min(len(angles), i + window_size // 2 + 1)
        window = angles[start_idx:end_idx]
        smoothed_angles.append(np.mean(window))
    return smoothed_angles

def issue_warning(frame, frame_counter):
    if (frame_counter // 5) % 2 == 0:  # Blink effect: display only on alternate frames
        height, width, _ = frame.shape
        warning_text = "WARNING: Pedestrian approaching!"
        text_size, _ = cv2.getTextSize(warning_text, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
        text_x = (width - text_size[0]) // 2
        text_y = text_size[1] + 10
        cv2.putText(frame, warning_text, (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1)

def draw_transparent_background(frame, x, y, text, bbox_height):
    # Get the size of the text
    font_scale = 0.4  # Smaller font scale
    (text_width, text_height), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, font_scale, 1)
    
    # Adjust font scale if text is wider than the bounding box
    if text_width > (bbox_height * 2):  # Adjust based on bounding box width
        font_scale = 0.3  # Further reduce font size
    
    # Recalculate text size with adjusted font scale
    (text_width, text_height), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, font_scale, 1)
    
    # Define the background rectangle coordinates
    bg_x1 = x - 5
    bg_y1 = y - text_height - 5
    bg_x2 = x + text_width + 5
    bg_y2 = y + 5
    
    # Create a transparent overlay
    overlay = frame.copy()
    cv2.rectangle(overlay, (bg_x1, bg_y1), (bg_x2, bg_y2), (0, 0, 0), -1)  # Black background
    alpha = 0.5  # Transparency factor
    cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)
    
    # Draw the text
    cv2.putText(frame, text, (x, y), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), 1)  # Thinner font

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.resize(frame, (720, 720))

    frame_counter += 1
    if frame_counter % 5 != 0:
        continue
    
    results = model(frame)
    beep_needed = False

    # Reset tracking if the number of vehicles exceeds the limit
    if len(prev_positions) > max_vehicles_to_track:
        prev_positions = {}

    # Sort vehicles by bounding box size (height) and keep only the closest 2-3
    vehicles = []
    for result in results:
        bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")
        classes = np.array(result.boxes.cls.cpu(), dtype="int")

        for bbox, cls in zip(bboxes, classes):
            (x, y, x2, y2) = bbox
            height = y2 - y
            vehicles.append((height, (x, y, x2, y2), cls))

    # Sort vehicles by height (largest first, assuming closer vehicles are larger)
    vehicles.sort(key=lambda x: x[0], reverse=True)
    vehicles = vehicles[:max_vehicles_to_track]  # Keep only the closest 2-3 vehicles

    for _, (x, y, x2, y2), cls in vehicles:
        centroid_x = (x + x2) // 2
        centroid_y = (y + y2) // 2
        height = y2 - y

        person_id = str(cls)
        prev_positions.setdefault(person_id, []).append((centroid_x, centroid_y))

        if len(prev_positions[person_id]) >= 2:
            prev_x, prev_y = prev_positions[person_id][-2]
            curr_x, curr_y = prev_positions[person_id][-1]
            horizontal_direction = "Right" if curr_x > prev_x else "Left"
            vertical_direction = "Backward" if curr_y < prev_y else "Forward"
            color = (0, 255, 255) if vertical_direction == "Forward" else (255, 0, 0)
            if height > BEEP_THRESHOLD:
                color = (0, 0, 255)
                beep_needed = True
            
            # Draw bounding box
            cv2.rectangle(frame, (x, y), (x2, y2), color, 1)  # Thinner bounding box
            
            # Draw transparent background and direction text
            direction_text = f"{vertical_direction}, {horizontal_direction}"
            draw_transparent_background(frame, x, y - 5, direction_text, height)  # Adjusted position
            
            if vertical_direction == "Forward" and horizontal_direction == "Right" and beep_active:
               issue_warning(frame, frame_counter)
    
    if beep_needed and not beep_active:
        beep_sound.play(-1)
        beep_active = True
    elif not beep_needed and beep_active:
        beep_sound.stop()
        beep_active = False
    
    cv2.imshow("ADAS PeVision", frame)
    if cv2.waitKey(300) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
pygame.quit()
```

```python
import cv2
import numpy as np
import pygame
import math
from ultralytics import YOLO

# Initialize pygame for sound
pygame.init()
beep_sound = pygame.mixer.Sound("beep.wav")

# Load YOLOv8 model
model = YOLO("yolov8m.pt")
cap = cv2.VideoCapture("data9.mov")

# Tracking and state management
prev_positions = {}
speed_records = {}  # To store previous positions for speed calculations
beep_active = False
frame_counter = 0
BEEP_THRESHOLD = 400 # Threshold for triggering the beep
max_vehicles_to_track = 4  # Maximum number of vehicles to track

def smooth_angles(angles, window_size=5):
    smoothed_angles = []
    for i in range(len(angles)):
        start_idx = max(0, i - window_size // 2)
        end_idx = min(len(angles), i + window_size // 2 + 1)
        window = angles[start_idx:end_idx]
        smoothed_angles.append(np.mean(window))
    return smoothed_angles

def issue_warning(frame, frame_counter):
    if (frame_counter // 5) % 2 == 0:  # Blink effect: display only on alternate frames
        height, width, _ = frame.shape
        warning_text = "WARNING: Vehicle approaching!"
        text_size, _ = cv2.getTextSize(warning_text, cv2.FONT_HERSHEY_SIMPLEX, 1.5, 3)
        text_x = (width - text_size[0]) // 2
        text_y = text_size[1] + 10
        cv2.putText(frame, warning_text, (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3)

def draw_transparent_background(frame, x, y, text, bbox_height):
    font_scale = 0.4
    (text_width, text_height), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, font_scale, 1)
    bg_x1 = x - 5
    bg_y1 = y - text_height - 5
    bg_x2 = x + text_width + 5
    bg_y2 = y + 5
    overlay = frame.copy()
    cv2.rectangle(overlay, (bg_x1, bg_y1), (bg_x2, bg_y2), (0, 0, 0), -1)
    alpha = 0.5
    cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)
    cv2.putText(frame, text, (x, y), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), 1)

def detect_free_space(frame):
    height, width, _ = frame.shape
    margin = 20  # Margin to keep the rectangle inside the frame borders
    
    # Define the region of interest (ROI) for free space detection
    roi = frame[margin:height - margin, margin:width - margin]
    
    # Convert ROI to grayscale and threshold
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)
    
    # Create a mask to exclude regions occupied by cars or obstacles
    mask = np.ones_like(thresh) * 255
    
    # Define regions to exclude (left, right, top, bottom)
    exclude_left = 100  # Width of the left region to exclude
    exclude_right = 100  # Width of the right region to exclude
    exclude_top = 50  # Height of the top region to exclude
    exclude_bottom = 50  # Height of the bottom region to exclude
    
    # Set the excluded regions to 0 in the mask
    mask[:, :exclude_left] = 0
    mask[:, -exclude_right:] = 0
    mask[:exclude_top, :] = 0
    mask[-exclude_bottom:, :] = 0
    
    # Apply the mask to the thresholded image
    thresh = cv2.bitwise_and(thresh, thresh, mask=mask)
    
    # Find contours in the thresholded image
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    max_area = 0
    best_contour = None
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > max_area:
            max_area = area
            best_contour = contour
    
    if best_contour is not None:
        # Get bounding box for the largest contour
        x, y, w, h = cv2.boundingRect(best_contour)
        
        # Adjust coordinates to account for the margin
        x += margin
        y += margin
        
        # Draw the free space rectangle
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame, "Free Space", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    frame_counter += 1
    if frame_counter % 5 != 0:
        continue
    
    results = model(frame)
    beep_needed = False
    
    if len(prev_positions) > max_vehicles_to_track:
        prev_positions = {}

    vehicles = []
    for result in results:
        bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")
        classes = np.array(result.boxes.cls.cpu(), dtype="int")
        confidences = np.array(result.boxes.conf.cpu(), dtype="float")

        for bbox, cls, conf in zip(bboxes, classes, confidences):
            (x, y, x2, y2) = bbox
            height = y2 - y
            vehicles.append((height, (x, y, x2, y2), cls, conf))
    
    vehicles.sort(key=lambda x: x[0], reverse=True)
    vehicles = vehicles[:max_vehicles_to_track]
    
    for _, (x, y, x2, y2), cls, conf in vehicles:
        centroid_x = (x + x2) // 2
        centroid_y = (y + y2) // 2
        height = y2 - y
        person_id = str(cls)
        
        prev_positions.setdefault(person_id, []).append((centroid_x, centroid_y))
        
        # Calculate speed
        if len(prev_positions[person_id]) >= 2:
            prev_x, prev_y = prev_positions[person_id][-2]
            curr_x, curr_y = prev_positions[person_id][-1]
            
            speed = math.sqrt((curr_x - prev_x) ** 2 + (curr_y - prev_y) ** 2)
            speed_text = f"Speed: {speed:.2f} px/frame"
            draw_transparent_background(frame, x, y - 35, speed_text, height)
            
            horizontal_direction = "Right" if curr_x > prev_x else "Left"
            vertical_direction = "Backward" if curr_y < prev_y else "Forward"
            color = (0, 255, 255) if vertical_direction == "Forward" else (255, 0, 0)
            
            if height > BEEP_THRESHOLD:
                color = (0, 0, 255)
                beep_needed = True
            
            cv2.rectangle(frame, (x, y), (x2, y2), color, 2)
            
            # Get class name from model.names
            class_name = model.names[int(cls)]
            class_text = f"{class_name}: {conf:.2f}"
            draw_transparent_background(frame, x, y - 5, class_text, height)
            
            direction_text = f"{vertical_direction}, {horizontal_direction}"
            draw_transparent_background(frame, x, y - 20, direction_text, height)
            
            if vertical_direction == "Forward" and horizontal_direction == "Right" and beep_active:
                issue_warning(frame, frame_counter)

    if beep_needed and not beep_active:
        beep_sound.play(-1)
        beep_active = True
    elif not beep_needed and beep_active:
        beep_sound.stop()
        beep_active = False
    
    detect_free_space(frame)  # Call free space detection function
    cv2.imshow("ADAS PeVision", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

angles = []
for person_id, trajectory in prev_positions.items():
    if len(trajectory) >= 3:
        for i in range(2, len(trajectory)):
            current_position = trajectory[i]
            previous_position = trajectory[i - 1]
            angle = math.atan2(current_position[1] - previous_position[1], current_position[0] - previous_position[0])
            angles.append(angle)

if angles:
    smoothed_angles = smooth_angles(angles)
    average_error_degrees = (sum(smoothed_angles) / len(smoothed_angles)) * (180 / np.pi)
    print("Smoothed Average angular error (degrees):", average_error_degrees)

cap.release()
cv2.destroyAllWindows()
pygame.quit()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO  # Ensure you have YOLOv8 installed

# Load YOLO model
model = YOLO("yolov8m.pt")  # Change to your model path if needed

# Open video file or webcam
cap = cv2.VideoCapture("data8.mp4")  # Change to 0 for webcam

if not cap.isOpened():
    print("Error: Could not open video.")
else:
    frame_counter = 0  # Initialize frame counter
    frame_save_count = 5  # Number of frames to display
    saved_frames = 0  # Counter for displayed frames

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret or saved_frames >= frame_save_count:
            break

        frame_counter += 1
        if frame_counter % 10 != 0:  # Process every 10th frame
            continue

        results = model(frame)

        for result in results:
            bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")
            classes = np.array(result.boxes.cls.cpu(), dtype="int")

            for bbox, cls in zip(bboxes, classes):
                (x, y, x2, y2) = bbox
                cv2.rectangle(frame, (x, y), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, f"Class {cls}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2)

        # Convert BGR (OpenCV) to RGB (Matplotlib)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        plt.figure(figsize=(6, 6))
        plt.imshow(frame_rgb)
        plt.axis("off")
        plt.show()

        saved_frames += 1

    cap.release()
```
